```python
from collections import Counter
import random as r 
```

```python

while True:
    l=['apple','orange','banana','melon','grape']
    secret=r.choice(l)
    letterguess=''
    chances=len(secret)+2
    flag=False

    print('guess the secret')
    for i in secret:
        print('_',end=' ')
    while flag==False and chances>0:
        print('chances left',chances)    
        guess=input('enter a letter to guess')
        if not guess.isalpha():
            print('enter only numbers')
            continue
        elif len(guess)>1:
            print('enter only one character')
            continue
        elif guess in letterguess:
            print('u already guessed this letter')
            continue
        chances=chances-1

        if guess in secret and guess not in letterguess:
            letterguess=letterguess+guess

        for char in secret:
            if char in letterguess:
                print(char, end = ' ')
            else:
                print('_',end=' ')
        for char in secret:
            if char not in letterguess:
                flag=False
                break
        else:
            flag=True
        if flag==True:
            print('\n you won the game the word is',secret)
            break
    if flag==False and chances==0:
        print('\n u lost the word was', secret)
    play=input('u want to play again?y/n')
    if play=='n':
        break
```
